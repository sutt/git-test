```python
projdir = myroot
projdir = os.path.normpath(projdir)
scripts = [os.path.join(r,x) 
           for r,s,f in os.walk(projdir) 
           for x in f]
```

```python
s = filter(lambda f: ".txt" in  f,scripts)
print s[:4]
len(s)
```

```python
ss = filter(lambda f: f[-4:] == ".txt", scripts)
print ss[:4]
len(ss)
```
